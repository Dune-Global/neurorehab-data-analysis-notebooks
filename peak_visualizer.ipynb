```python
def find_peaks(arr):
    difference = np.diff(arr)
    nonzero_difference = difference[difference != 0]
    signs = np.sign(nonzero_difference)
    peaks = np.where((np.diff(np.sign(signs)) == -2) & (np.sign(signs[:-1]) == 1))[0] + 1
    return peaks

peak_indices = find_peaks(df['readingValue'].values)
peak_values = df['readingValue'].values[peak_indices]

print(f"Peak Indices: {peak_indices}")
print(f"Peak Values: {peak_values}")
```

```python
from sklearn.ensemble import IsolationForest

def detect_anomalies(arr, contamination=0.01):
    clf = IsolationForest(contamination=contamination)
    clf.fit(arr.reshape(-1, 1))
    scores = clf.decision_function(arr.reshape(-1, 1))
    labels = clf.predict(arr.reshape(-1, 1))
    return labels, scores

labels, scores = detect_anomalies(df['readingValue'].values)

anomalies = df[labels == -1]

print(f"Number of anomalies: {len(anomalies)}")
```

```python
plt.plot(df['readingNumber'], df['readingValue'])
plt.scatter(anomalies['readingNumber'], anomalies['readingValue'], color='red')
plt.xlabel("Reading Number")
plt.ylabel("Reading Value")
plt.title("Reading Values Over Time with Anomalies")
plt.grid(True)
plt.show()
```

```python
def find_valleys(arr):
    difference = np.diff(arr)
    nonzero_difference = difference[difference != 0]
    signs = np.sign(nonzero_difference)
    valleys = np.where((np.diff(np.sign(signs)) == 2) & (np.sign(signs[1:]) == 1))[0] + 1
    return valleys
```

```python
valley_indices = find_valleys(y_values)
valley_values = y_values[valley_indices]

print(f"Valley Indices: {valley_indices}")
print(f"Valley Values: {valley_values}")
```
